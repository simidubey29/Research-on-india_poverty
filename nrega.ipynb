```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("NREGA.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.rename(columns={
    'Total No. of Workers': 'Total_Workers',
    'Total No. of Active Workers': 'Active_Workers',
    'Women Persondays': 'Women_Persondays',
    'Average days of employment provided per Household': 'Avg_Employment_Days',
    'Average Wage rate per day per person(Rs.)': 'Avg_Wage'
}, inplace=True)

```

```python
state_df = df.groupby("state_name").agg({
    "Total_Workers": "sum",
    "Active_Workers": "sum",
    "Women_Persondays": "sum",
    "Avg_Employment_Days": "mean",
    "Avg_Wage": "mean"
}).reset_index()

state_df.head()
```

```python
plt.figure()
sns.barplot(data=state_df.sort_values("Total_Workers", ascending=False).head(10),
            x="Total_Workers", y="state_name")
plt.title("Top 10 States by Total Workers (MGNREGA)")
plt.show()
```

```python
plt.figure()
sns.barplot(data=state_df.sort_values("Avg_Employment_Days", ascending=False),
            x="Avg_Employment_Days", y="state_name")
plt.title("Average Employment Days per Household")
plt.show()
```

```python
plt.figure()
sns.barplot(data=state_df.sort_values("Avg_Wage", ascending=False),
            x="Avg_Wage", y="state_name")
plt.title("Average Wage Rate by State")
plt.show()
```

```python
plt.figure()
sns.barplot(data=state_df.sort_values("Women_Persondays", ascending=False).head(10),
            x="Women_Persondays", y="state_name")
plt.title("Top States by Women Persondays")
plt.show()
```

```python
plt.figure()
sns.heatmap(state_df[['Total_Workers','Active_Workers',
                      'Avg_Employment_Days','Avg_Wage']].corr(),
            annot=True)
plt.title("Correlation Between Poverty Indicators")
plt.show()
```

```python
plt.figure()
sns.histplot(df['Total_Workers'], bins=30, kde=True)
plt.title("Distribution of Total Workers under MGNREGA")
plt.xlabel("Total Workers")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Avg_Employment_Days'])
plt.title("Average Employment Days per Household")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Avg_Employment_Days', y='Avg_Wage', data=df)
plt.title("Employment Days vs Wage Rate")
plt.xlabel("Avg Employment Days")
plt.ylabel("Avg Wage (â‚¹)")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Total_Workers', y='Active_Workers', data=df)
plt.title("Total Workers vs Active Workers")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Women_Persondays', y='Avg_Employment_Days', data=df)
plt.title("Women Participation vs Employment Days")
plt.show()
```

```python
state_df = df.groupby("state_name").agg({
    "Total_Workers": "sum",
    "Women_Persondays": "sum",
    "Avg_Employment_Days": "mean",
    "Avg_Wage": "mean"
}).reset_index()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(
    data=state_df.sort_values("Total_Workers", ascending=False).head(10),
    x="Total_Workers", y="state_name"
)
plt.title("Top 10 States by MGNREGA Workers")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(
    state_df[['Total_Workers','Avg_Employment_Days','Avg_Wage','Women_Persondays']].corr(),
    annot=True, cmap="coolwarm"
)
plt.title("Correlation between Poverty Indicators")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=state_df,
    x="Avg_Employment_Days",
    y="Avg_Wage",
    size="Total_Workers",
    sizes=(50, 1000),
    legend=False
)
plt.title("Employment, Wage & Poverty Dependency (Bubble Plot)")
plt.show()
```
